```python
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
df=pd.read_csv('Salary Data.csv')
df
```

```python
plt.scatter(df['Exp_year'],df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Salary based on exp_years')
```

```python
x=df[['Exp_year']]
x
```

```python
y=df[['Salary']]
y
```

```python
#descriptive statistics
df.describe()
```

```python
xmean=x.mean()
ymean=y.mean()
print(xmean)
print(ymean)
```

```python
plt.scatter(df['Exp_year'],df['Salary'])
plt.scatter(x.mean(),y.mean(),color='green') #mean value of x and y
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Salary based on exp_years')
```

```python
#importing lineargression from scikit learn
from sklearn.linear_model import LinearRegression
```

```python
#creating object
reg=LinearRegression()
```

```python
#training the model
reg.fit(x,y)
```

```python
#y=mx+c, m=coefficient, c=intercept of y
m=reg.coef_
c=reg.intercept_
print(m,c)
```

```python
#predicting salary for 12 years of experienced employee
py=m*12+c
py
```

```python
df['Predicted']=reg.predict(x)
df
```

```python
plt.plot(x,reg.predict(x),color='black') #linear line
plt.scatter(df['Exp_year'],df['Salary']) #represented by blue color
plt.scatter(df['Exp_year'],reg.predict(x),color='red') #predicted value
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Salary based on exp_years')
```

```python
from sklearn.metrics import r2_score
```

```python
#accuracy or performance of the model
r2_score(y,reg.predict(x))*100
```

```python
#another way to determine the performance
accuracy=reg.score(x,y)
print(accuracy*100)
```
